```python
principals= "C:\\Users\\559816\\Documents\\Project2\\Projet-Data-IA\\Data\\movies_title_principals_cleaned.csv"
df_princip = pd.read_csv(principals, sep = ',')
```

```python
df_princip = df_princip[df_princip['category'].isin(['actress', 'actor', 'director'])]
```

```python
nbasics = "C:\\Users\\559816\\Documents\\Project2\\Projet-Data-IA\\Data\\cleaned_name_basics.csv"
df_nbasics = pd.read_csv(nbasics, sep = ',')
```

```python
names = pd.merge(df_nbasics,
                  df_princip,
                  how = "left",
                  left_on = "nconst",
                  right_on = "nconst")
```

```python
names = names.drop(columns = ['primaryProfession', 'knownForTitles'])
```

```python
names_directors = names[names['category'].isin(['director'])]
names_directors = names_directors.drop(columns = ['tconst', 'category'])
names_directors = names_directors.drop_duplicates()
```

```python
merged = pd.merge(merged,
                  names_directors,
                  how = "left",
                  left_on="directors",
                  right_on="nconst")
```

```python
merged = merged.rename(columns={'primaryName': 'Director_name'})
merged = merged.drop(columns = ['nconst'])
```

```python
names_actors = names[names['category'].isin(['actress', 'actor'])]
names_actors = names_actors.drop(columns = ['nconst'])
```

```python
names_actors = names_actors.groupby('tconst')['primaryName'].agg(list)
```

```python
merged = pd.merge(merged,
                  names_actors,
                  how = "left",
                  left_on = "titleId",
                  right_on = "tconst")
```

```python
merged = merged.drop(columns = ['directors'])
merged = merged.rename(columns={'primaryName': 'Actors_Actresses'})

merged.head()
```

```python
merged['backdrop_path'].fillna(merged['poster_path'], inplace = True)
```

```python
merged = merged.dropna()

merged.head()
```

```python
merged.to_csv('merged_all_tables.csv', index=False) 
```

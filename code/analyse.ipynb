```python
def clean_actors_list(actors_list):
    unique_actors = sorted(set(actors_list))
    cleaned_list = ', '.join(unique_actors)
    return cleaned_list

df['Actors_Actresses'] = df['Actors_Actresses'].apply(eval).apply(clean_actors_list)
df['production_companies_name'] = df['production_companies_name'].apply(eval).apply(clean_actors_list)
df
```

```python
df.to_csv('merged_final.csv', index=False) 
```

```python
import matplotlib.pyplot as plt

unique_distribution = df['genres'].value_counts()
unique_distribution.plot(kind='bar')
plt.xlabel('Unique Values')
plt.ylabel('Frequency')
plt.title('Distribution of Unique Values in Column')

plt.show()
```

```python
startYear_distribution = df['startYear'].value_counts()

plt.bar(startYear_distribution.index, startYear_distribution.values)
plt.xlabel('Unique Values')
plt.ylabel('Qty')
plt.title('Distribution of Unique Values in startYear Column')

plt.show()
```

```python
Director_name_distribution = df['Director_name'].value_counts()

top_directors = Director_name_distribution.head(20)

plt.bar(top_directors.index, top_directors.values)
plt.xlabel('Director Name')
plt.ylabel('Frequency')
plt.title('Distribution of Director Names')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlap of labels
plt.show()
```

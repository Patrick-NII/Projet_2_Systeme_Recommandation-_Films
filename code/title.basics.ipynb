```python
import pandas as pd
```

```python
movies_title_basics = "C:\\Users\\559816\\Documents\\Project2\\Projet-Data-IA\\Database\\not cleaned\\movies_title_basics.csv"
df = pd.read_csv(movies_title_basics, sep = ',')
```

```python
df.head()
```

```python
import numpy as np

df.replace(r'\N', np.nan, inplace=True)

df.head()
```

```python
df = df.drop(columns=['titleType', 'endYear', 'primaryTitle', 'originalTitle', 'isAdult'])
df.head()
```

```python
df['genres'].unique()
```

```python
df = df.loc[df["genres"].isin(['Romance', 'Documentary', 'Action',
       'Adventure', 'Biography', 'Drama', 'Fantasy', 'Comedy', 'War',
       'Crime', 'Family', 'History', 'Sci-Fi', 'Thriller', 'Western',
       'Mystery', 'Horror', 'Animation', 'Musical', 'Film-Noir'])]

df.head()
```

```python
df['genres'].nunique()
```

```python
df['genres'] = df['genres'].str.split(',')
df = df.explode('genres')
df.reset_index(drop=True, inplace=True)

df.head()
```

```python
# ready to export

df.to_csv('cleaned_movies_title_basics.csv', index=False)
```
